# cos wave

```python
import numpy as np
import matplotlib.pyplot as plt
t = np.arange(0.0,2.0,0.01)
s = 1 + np.cos(2*np.pi*t)
plt.plot(t,s,"*")
plt.grid()
plt.xlabel('Time (t)')
plt.ylabel('Voltage (mv)')
plt.title('Cosine wave plot()')
```

# subplots

```python
x1 = np.linspace(0.,0.5)
x2 = np.linspace(0.0,2.0)
y1 = np.cos(2*np.pi*x1)*np.exp(-x1)
y2 = np.cos(2*np.pi*x2)
plt.subplot(2,1,1)
plt.plot(x1,y1,'--')
plt.title('Subplot - 1')
plt.xlabel('x1')
plt.ylabel('Amp(y1)')
plt.subplot(2,1,2)
plt.plot(x1,y1,'*')
plt.title('Subplot - 2')
plt.xlabel('x2')
plt.ylabel('Amp(y2)')
plt.show()
```

# Bar Graph

```python
x = [1,2,3,4,5]
y = [10,24,36,40,5]
tick_labels = ['one','two','three','four','five']
plt.bar(x,y,tick_label=tick_labels,width=0.3,color=['green','blue','pink'])
plt.xlabel(" x - axis")
plt.ylabel(" y - axis")
plt.title('Bar Graph')
plt.show()
```
